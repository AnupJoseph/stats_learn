```
# Data science imports
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```
college_data = pd.read_csv("https://raw.githubusercontent.com/nguyen-toan/ISLR/master/dataset/College.csv")
college_data.columns.values[0] = "College names"
college_data
```

```
college_data.describe()
```

```
index_vals = college_data["Private"].astype("category").cat.codes
dimensions = [dict(label=col,values=college_data[col]) for i,col in enumerate(college_data.iloc[:,2:12])]
```

```
fig = go.Figure()
fig.add_trace(
    go.Splom(dimensions = dimensions,text=college_data["Private"],marker=dict(color=index_vals,showscale=False, # colors encode categorical variables
                            line_color='white', line_width=0.5)
                ))

fig.update_layout(
    title='USA collges Data set',
    width=1000,
    height=1000,
)
fig.show()
```

```
fig = go.Figure()
fig.add_trace(go.Box(y=college_data["Outstate"],name="Outstate"))
fig.add_trace(go.Box(y=college_data["Private"],name="Private"))
fig.show()
```

```
college_data["Elite"] = college_data["Top10perc"].apply(lambda x:x>50) 
college_data["Elite"]
```
